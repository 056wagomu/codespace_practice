```python
import pandas as pd
import pickle
pd.set_option('display.max_columns', 500)

SC_data = pd.read_csv("/workspaces/codespace_practice/codespace_test_module/final_SC_X_data_ver2.csv")
X_data = pd.concat([SC_data.loc[:, :"composition of MA10"], SC_data.loc[:, "value of oxygen"], SC_data.loc[:, "Tc"]], axis = "columns")
```

```python
display(X_data)
```

```python
round_list = ["composition of MA1", "composition of MA2", "composition of MA3", "composition of MA4", "composition of MA5", "composition of MA6", "composition of MA7", "composition of MA8", "composition of MA9", "composition of MA10"]
for col in round_list:
    X_data.loc[:, col] = X_data.loc[:, col].round().fillna(0)
    X_data[col] = X_data[col].astype(int)
    
X_data.loc[:, "value of oxygen"] = X_data.loc[:, "value of oxygen"].round().fillna(0)
X_data["value of oxygen"] = X_data["value of oxygen"].astype(int)
```

```python
X_data.loc[:, "new_formula"] = ""

for j in range(len(X_data)):
    for i in range(5, 23, 2):
        if not(pd.isnull(X_data.iloc[j, i])) and X_data.iloc[j, i+1] != 0:
            X_data.loc[j, "new_formula"] += X_data.iloc[j, i] + str(X_data.iloc[j, i+1])
    
    if X_data.loc[j, "value of oxygen"] != 0:
        X_data.loc[j, "new_formula"] += "O" + str(X_data.loc[j, "value of oxygen"])
```

```python
display(X_data.loc[X_data["new_formula"] == "", :])
```

```python
#ここで成型したdfをver3として出力して保存
X_data.to_csv("/workspaces/codespace_practice/codespace_test_module/final_SC_X_data_ver3.csv", index = False)
```
