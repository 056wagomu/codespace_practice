```python
import pandas as pd
X_data = pd.read_csv("/workspaces/codespace_practice/codespace_test_module/final_SC_X_data_ver2.csv")
X_data = pd.concat([X_data.loc[:, :"composition of MA10"], X_data.loc[:, "value of oxygen"]], axis = "columns")
```

```python
pd.set_option('display.max_columns', 500)
```

```python
display(X_data.head())
```

```python
round_list = ["composition of MA1", "composition of MA2", "composition of MA3", "composition of MA4", "composition of MA5", "composition of MA6", "composition of MA7", "composition of MA8", "composition of MA9", "composition of MA10"]
for col in round_list:
    X_data.loc[:, col] = X_data.loc[:, col].round().fillna(0)
    X_data.loc[:, col] = X_data.loc[:, col].astype(int)
```

```python
display(X_data.head())
```

```python
#もし物質名の列が欠損値でなく比が0でない場合に足し合わせて組成式を作る
X_data.loc[:, "new_formula"] = ""

for i in range(5, 23, 2):
    for j in range(len(X_data)):
        if not(pd.isnull(X_data.iloc[j, i])) and X_data.iloc[j, i+1] != 0:
            X_data.loc[j, "new_formula"] += X_data.iloc[j, i] + str(int(X_data.iloc[j, i+1]))
```
